```python
import os
import json
import pandas as pd

# Define the directory where the performance data is saved
PERFORMANCE_SAVE_DIR = '/content/drive/MyDrive/Colab Notebooks/performance_data'

# Create a list to store the performance data from each model
all_performance_data = []

# Iterate through the files in the directory
for filename in os.listdir(PERFORMANCE_SAVE_DIR):
    if filename.endswith(".json"):
        filepath = os.path.join(PERFORMANCE_SAVE_DIR, filename)
        try:
            with open(filepath, 'r') as f:
                data = json.load(f)
                all_performance_data.append(data)
        except Exception as e:
            print(f"Error reading file {filename}: {e}")

# Convert the list of dictionaries into a pandas DataFrame
performance_df = pd.DataFrame(all_performance_data)

# Display the DataFrame
print("\nModel Performance Summary:")
display(performance_df)
```

```python
# Define the directory where the resource usage data is saved
RESOURCE_USAGE_SAVE_DIR = '/content/drive/MyDrive/Colab Notebooks/model_resource_usage_logs'

# Create a list to store the resource usage data from each model
all_resource_usage_data = []

# Iterate through the files in the directory
if os.path.exists(RESOURCE_USAGE_SAVE_DIR):
    for filename in os.listdir(RESOURCE_USAGE_SAVE_DIR):
        # Change from .json to .csv
        if filename.endswith(".csv"):
            filepath = os.path.join(RESOURCE_USAGE_SAVE_DIR, filename)
            try:
                # Use pd.read_csv to read CSV files
                data = pd.read_csv(filepath)
                all_resource_usage_data.append(data)
            except Exception as e:
                print(f"Error reading file {filename}: {e}")

    # Concatenate the list of DataFrames into a single DataFrame
    if all_resource_usage_data:
        resource_usage_df = pd.concat(all_resource_usage_data, ignore_index=True)

        # Display the DataFrame
        print("\nModel Resource Usage Summary:")
        display(resource_usage_df)
    else:
        print("No CSV files found in the directory.")
else:
    print(f"Directory not found: {RESOURCE_USAGE_SAVE_DIR}")
```
